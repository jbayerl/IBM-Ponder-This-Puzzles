<h2>Algorithm</h2>

Start with last value of array (of length n+1). 

1. If C then we want C -> T -> G:

2. If A then we want A -> C then do 1.

3. If T then we want T -> G,
    




 ->C Functions

```python
def T_C(arr,i):
    '''Takes as input the array and index position to transform'''
    global count
    global result
    if arr[i] != 'T':                                       
        print('Error in T_C function at index',i)
    else:
        '''We can do anything with arr[0]'''
        if i == 0:                                                   
            arr[i] = 'C'
            #print(arr,count)
            count +=1
            return arr
        else:
            '''If (i-1)-th position is C we repeat with arr[0:i-1]'''
            if arr[i-1] == 'C':
                for j in range(i-1,-1,-1):
                    if arr[j] =='T':
                        T_C(arr,j)
                 
                    if arr[j] =='G':
                        T_C(G_T(arr,j),j)
             
                    if arr[j] =='A':
                        A_C(arr,j)
                
                
            else:        
                if i-1 == 0:
                    arr[i-1] = 'C'
                    #print(arr,count)
                    count +=1
                else:
                    if arr[i-1] =='T':
                        T_C(arr,i-1)

                    if arr[i-1] =='G':
                        T_C(G_T(arr,i-1),i-1)

                    if arr[i-1] =='A':
                        A_C(arr,i-1)
           
            arr[i] = 'C'
            #print(arr,count)
            count +=1
            return arr
           

def A_C(arr,i):
    '''Takes as input the array to transform'''
    global count
    if arr[i] != 'A':                                       
        print('Error in A_C function at index',i)
    else:
        '''We can do anything with arr[0]'''
        if i == 0:                                                   
            arr[i] = 'C'
            #print(arr,count)
            count +=1
            return arr
        else:
            '''If (i-1)-st is C we continue with turning the remaining entries to A'''
            if arr[i-1] == 'C':
                for j in range(i-2,-1,-1):
                    if arr[j] =='T':
                        C_A(T_C(arr,j),j)
                  
                    if arr[j] =='G':
                        C_A(T_C(G_T(arr,j),j),j)
                 
                    if arr[j] =='C':
                        C_A(arr,j)
                 
            else:        
                if i-1 == 0:
                    arr[i-1] = 'C'
                    #print(arr,count)
                    count +=1
                    return A_C(arr,i)
                else:
                    if arr[i-1] =='T':
                        A_C(T_C(arr,i-1),i)
                    if arr[i-1] =='G':
                        A_C(T_C(G_T(arr,i-1),i-1),i)
                    if arr[i-1] =='A':
                        A_C(A_C(arr,i-1),i)
                        
            if arr[i] != 'C':           
                arr[i]='C'
                #print(arr,count)
                count +=1
            return arr
        
```

```python
def C_T(arr,i):
    '''Takes as input the array and index position to transform'''
    global count
    if arr[i] != 'C':                                       
        print('Error in C_T function at index',i)
    else:
        '''We can do anything with arr[0]'''
        if i == 0:                                                   
            arr[i] = 'T'
            #print(arr,count)
            count +=1
            return arr
        else:
            '''If (i-1)-th position is T we repeat with arr[0:i-1]'''
            if arr[i-1] == 'T':
                for j in range(i-2,-1,-1):
                    if arr[j] =='A':
                        C_T(A_C(arr,j),j)
                  
                    if arr[j] =='G':
                        G_T(arr,j)
                  
                    if arr[j] =='C':
                        C_T(arr,j)
            else:        
                if i-1 == 0:
                    arr[i-1] = 'T'
                    #print(arr,count)
                    count +=1
                    
                else:
                    if arr[i-1] =='C':
                        C_T(arr,i-1)

                    if arr[i-1] =='G':
                        G_T(arr,i-1)

                    if arr[i-1] =='A':
                        C_T(A_C(arr,i-1),i-1)     
                      
            arr[i] = 'T'
            #print(arr,count)
            count +=1
            return arr


def G_T(arr,i): #we don't actually need this but OK 
    '''Takes as input the array to transform'''
    global count
    if arr[i] != 'G':                                       
        print('Error in G_T function at index',i)
    else:
        '''We can do anything with arr[0]'''
        if i == 0:                                                   
            arr[i] = 'T'
            #print(arr,count)
            count +=1
            return arr
        else:
            '''If (i-1)-st is T we continue with turning the remaining entries to T'''
            if arr[i-1] == 'T':
                for j in range(i-2,-1,-1):
                    if arr[j] =='A':
                        C_T(A_C(arr,j),j)
                    
                    if arr[j] =='G':
                        G_T(arr,j)
                      
                    if arr[j] =='C':
                        C_T(arr,j)
                      
            else:        
                if i-1 == 0:
                    arr[i-1] = 'T'
                    #print(arr,count)
                    count +=1
                else:
                    if arr[i-1] =='A':
                        C_T(A_C(arr,i-1),i-1)

                    if arr[i-1] =='G':
                        G_T(arr,i-1)

                    if arr[i-1] =='C':
                        C_T(arr,i-1)
                    
            arr[i]='T'
            #print(arr,count)
            count +=1
            return arr
            
                            
                        
                        
                        
           
```

T->G Function

```python
def T_G(arr,i):
    '''Takes as input the array to transform'''
    global count
    if arr[i] != 'T':                                       
        print('Error in T_G function at index',i)
    else:
        '''We can do anything with arr[0]'''
        if i == 0:                                                   
            arr[i] = 'G'
            #print(arr,count)
            count +=1
            return arr
        else:
            '''If (i-1)-st is C we continue with turning the remaining entries to A'''
            if arr[i-1] == 'C':
                for j in range(i-2,-1,-1):
                    if arr[j] =='T':
                        C_A(T_C(arr,j),j)
                        
                    if arr[j] =='G':
                        C_A(T_C(G_T(arr,j),j),j)
                        
                    if arr[j] =='C':
                        C_A(arr,j)
                
                arr[i]='G'
                #print(arr,count)
                count +=1
                return arr   
               
            else:        
                if i-1 == 0:
                    arr[i-1] = 'C'
                    #print(arr,count)
                    count +=1
                    return T_G(arr,i)
                else:
                    if arr[i-1] =='A':
                        T_G(A_C(arr,i-1),i)
                    if arr[i-1] =='G':
                        T_G(T_C(G_T(arr,i-1),i-1),i)
                    if arr[i-1] =='T':
                        T_G(T_C(arr,i-1),i)
            
            
                            
                        
```

C->A Function

```python
def C_A(arr,i):
    '''Takes as input the array to transform'''
    global count
    if arr[i] != 'C':                                       
        print('Error in C_A function at index',i)
    else:
        '''We can do anything with arr[0]'''
        if i == 0:                                                   
            arr[i] = 'A'
            #print(arr,count)
            count +=1
            return arr
        else:
            '''If (i-1)-st is C we continue with turning the remaining entries to A'''
            if arr[i-1] == 'C':
                for j in range(i-2,-1,-1):
                    if arr[j] =='T':
                        C_A(T_C(arr,j),j)
                        
                    if arr[j] =='G':
                        C_A(T_C(G_T(arr,j),j),j)
                    
                    if arr[j] =='C':
                        C_A(arr,j)
            else:        
                if i-1 == 0:
                    arr[i-1] = 'C'
                    #print(arr,count)
                    count +=1
                    return C_A(arr,i)
                else:
                    if arr[i-1] =='A':
                        C_A(A_C(arr,i-1),i)
                    if arr[i-1] =='G':
                        C_A(T_C(G_T(arr,i-1),i-1),i)
                    if arr[i-1] =='T':
                        C_A(T_C(arr,i-1),i)
                        
            if arr[i] != 'A':            
                arr[i]='A'
                #print(arr,count)
                count +=1
            return arr
            
```
